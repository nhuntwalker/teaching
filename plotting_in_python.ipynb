This is a brief tutorial of how to make single-panel and multi-panel plots using Python.

### Import necessary packages

<b>Numpy</b> will allow us to make and use arrays, which make visualizing and manipulating data simple. We use the alias of "np" instead of the full name because it's easier than typing out "numpy" every time we want to use a function or module from numpy. We also use "np" instead of just importing all of numpy's functions and modules because we want to be able to differentiate between numpy's components and Python's pre-packaged components.

<b>Matplotlib</b> will allow us to make plots. We specifically use "pyplot" as that has all of the plotting functions that we'll need here. Note: you can do many of the same operations by importing Pylab. I just happen to use matplotlib.

```python
import numpy as np
import matplotlib.pyplot as plt
```

Any plot that we make using plt will pop open a separate window to display our data. I prefer to have all of my plots displayed in this window right here. As such, I'm going to use the magic function %matplotlib inline.  This will ensure that any cell of code that plots data will produce a plot of that data directly below.

```python
%matplotlib inline
```

### Read and Organize Your Data

Alright we're set to go.  Let's say that I have some data on the positions of some stars in our galaxy, and I want to visualize that data. I'll have to read that data in and plot it!  Let's read that data using Numpy's function, loadtxt( ). With np.loadtxt, you must specify the name of the file that you want to read in, and it has to be a string. Meaning, whatever you point np.loadtxt to has to have quotations around it.  Luckily, you can pass a variable that contains a string!

```python
input_file = 'mygalaxy.dat'

mydata = np.loadtxt(input_file)
```

<b>Note</b>: When reading data using np.loadtxt, your data must all have the same format. It will throw an error if there's a mix of floats (numbers) and strings (words). If you have some columns that are all numbers and some columns that are all words, use np.genfromtxt or open(input_file).readlines( ) instead. If you have individual columns that are a mix of words and numbers, use open(input_file).readlines( ). You will, however, have to do some extra work to make that information usable.

When np.loadtxt reads in data, it reads it in line by line. Our file, however, is formatted such that each line has 2 columns of separate position values (either galactic longitude or latitude). So, we have to make numpy go from showing storing our data line by line into storing our data column by column.

```python
gal_long, gal_lat = np.loadtxt(input_file, unpack=True)
```

### Plot Your Data

Let's start with a basic scatter plot. We want galactic longitude on the X axis, and latitude on the Y axis.

```python
plt.scatter(gal_long, gal_lat)
```

If you just needed to figure out how to plot data, you could stop right here. That's it. Your data is now plotted.  However, it looks ugly, and it doesn't really give you a good sense of what you're working with in your data. Let's finesse it up a little bit.

When making a scatter plot, there's 3 things you'll want to concern yourself with:

- point size
- point color
- point edge color

You can control each of these using key words in plt.scatter( ). Let's alter this plot such that our points are small, blue, and don't have black edges as follows:

```python
plt.scatter(gal_long, gal_lat, s=1, c='blue', edgecolor='None')
```

You can also change the point type by using the keyword "marker".  The available marker types are...

- 's': Square
- 'o': Circle
- 'D': Diamond
- '.': Point
- '^','v','>','<': Triangles
- '*': Star
- 'p': Pentagon
- '+': Plus signs
- 'x': X's
- '|': Vertical Lines
- '_': Horizontal Lines

The next thing to think about is: how will anyone know what I'm showing them when they look at the plot? In order to actually convey information, you'll need to label your axes. As a rule of thumb, <b>NEVER EVER EVER EVER EVER MAKE A PLOT WITHOUT LABELING YOUR AXES!!!</b>.  

For this we use plt.xlabel( ) and plt.ylabel( ). The sequence doesn't really matter here, just as long as they're in the same cell. For the hell of it, let's use plt.title( ) to label our plot as well.

```python
plt.title('My Galactic Latitude - Longitude Plot')
plt.scatter(gal_long, gal_lat, s=1, c='blue', edgecolor='None')
plt.xlabel('Galactic Longitude (degrees)')
plt.ylabel('Galactic Latitude (degrees)')
```

Next is the range of the data being plotted. Notice above that longitude only extends from -180 to +180 degrees, and latitude from -90 to +90.  As such, there's no need to extend past those limits. 

Let's use plt.xlim( ) and plt.ylim( ) to limit the window of what we plot.  Note that we can use these to zoom in to or out of different parts of data as well.

```python
plt.title('My Galactic Latitude - Longitude Plot')
plt.scatter(gal_long, gal_lat, s=1, c='blue', edgecolor='None')
plt.xlabel('Galactic Longitude (degrees)')
plt.ylabel('Galactic Latitude (degrees)')
plt.xlim(-180,180)
plt.ylim(-90,90)
```

When we changed our graph limits, we changed the scale of our axes! We can do a couple things to alter our axes so that they provide more information about the data:

- use minor tickmarks
- use different axis divisions
- use a grid

The first one is easy.  Just use plt.minorticks_on( )

```python
plt.title('My Galactic Latitude - Longitude Plot')
plt.scatter(gal_long, gal_lat, s=1, c='blue', edgecolor='None')
plt.xlabel('Galactic Longitude (degrees)')
plt.ylabel('Galactic Latitude (degrees)')
plt.xlim(-180,180)
plt.ylim(-90,90)
plt.minorticks_on()
```

The second one is a little more complicated.  My X-limits are currently going from -180 to 180, but only show ticks every 50 degrees.  Let's show ticks every 10 degrees.  First, we must declare an array of numbers.  This array will have every tick we want to see in our plot.

```python
dticks = 30
my_xticks = np.arange(-180, 180+dticks, dticks)

plt.title('My Galactic Latitude - Longitude Plot')
plt.scatter(gal_long, gal_lat, s=1, c='blue', edgecolor='None')
plt.xlabel('Galactic Longitude (degrees)')
plt.ylabel('Galactic Latitude (degrees)')
plt.xlim(-180,180)
plt.ylim(-90,90)
plt.xticks(my_xticks)
plt.minorticks_on()
```

The last one is as easy as the first one. However, I don't use it often because it makes my plot look really busy.  All you have to do is include plt.grid( ). By default, it'll plot dotted grid lines for each major tick on each axis.  You can choose the axes you want grids for with the keyword 'axis'.  You can have axis = 'x', 'y', or 'both'.

```python
dticks = 30
my_xticks = np.arange(-180, 180+dticks, dticks)

plt.title('My Galactic Latitude - Longitude Plot')
plt.scatter(gal_long, gal_lat, s=1, c='blue', edgecolor='None')
plt.xlabel('Galactic Longitude (degrees)')
plt.ylabel('Galactic Latitude (degrees)')
plt.xlim(-180,180)
plt.ylim(-90,90)
plt.xticks(my_xticks)
plt.minorticks_on()
plt.grid()
```

That X-axis is starting to get crowded. We can control the size of the figure and spread that X-axis out using plt.figure(figsize=(xsize,ysize)).

```python
dticks = 30
my_xticks = np.arange(-180, 180+dticks, dticks)

plt.figure(figsize=(10,4))
plt.title('My Galactic Latitude - Longitude Plot')
plt.scatter(gal_long, gal_lat, s=1, c='blue', edgecolor='None')
plt.xlabel('Galactic Longitude (degrees)')
plt.ylabel('Galactic Latitude (degrees)')
plt.xlim(-180,180)
plt.ylim(-90,90)
plt.xticks(my_xticks)
plt.minorticks_on()
plt.grid()
```

### Putting Text on a Figure

This part can be a major pain, which is why it gets its own section.  I want to put text on my plot.  Simple enough in concept.  For example, let's point out the Magellanic Clouds and the Galactic Bulge.  We use plt.text( xcoordinate, ycoordinate, textstring) to apply text to a graph.  You can change the font size with the keyword "fontsize", and the color with "color".

```python
dticks = 30
my_xticks = np.arange(-180, 180+dticks, dticks)

plt.figure(figsize=(10,4))
plt.title('My Galactic Latitude - Longitude Plot')
plt.scatter(gal_long, gal_lat, s=1, c='blue', edgecolor='None')
plt.text(-90, -50, 'LMC', fontsize=20)
plt.text(-60, -60, 'SMC', color='red')
plt.text(0, -30, 'MW Bulge')

plt.xlabel('Galactic Longitude (degrees)')
plt.ylabel('Galactic Latitude (degrees)')
plt.xlim(-180,180)
plt.ylim(-90,90)
plt.xticks(my_xticks)
plt.minorticks_on()
plt.grid()
```

This is all well and good, but plots are often not that simple. Perhaps I want to label this plot as part of a series of other plots (I'll get to multiple plots in a bit), and want all labels to be in the top left corner.  If I were to follow the above example, I'd have to put in very specific coordinates corresponding to the coordinate system for each plot. And if I ever changed the range of either axis, I'd have to revise those coordinates.

That's stupid.  We should be able to put text on a plot with axes that are relative to the plot itself. This will change some other things about how we plot data, but ultimately we're better for it. We have to use the keyword "transform", and initialize our plot a little differently.

```python
plt.figure(figsize=(10,4))
ax = plt.subplot(111)
ax.set_title('My Galactic Latitude - Longitude Plot')
ax.scatter(gal_long, gal_lat, s=1, c='blue', edgecolor='None')
ax.text(-90, -50, 'LMC', fontsize=20)
ax.text(-60, -60, 'SMC', color='red')
ax.text(0, -30, 'MW Bulge')
ax.text(0.03,0.9,'This is plot 1',transform=ax.transAxes)

ax.set_xlabel('Galactic Longitude (degrees)')
ax.set_ylabel('Galactic Latitude (degrees)')
ax.set_xlim(-180,180)
ax.set_ylim(-90,90)
ax.set_xticks(my_xticks)
ax.minorticks_on()
ax.grid()
```

### Multiple Plots

We have our beautiful plot of the galaxy. Let's say I want a separate plot of each quadrant of this data.  This calls for subplots (sorta alluded to earlier with ax = plt.subplot( )). When we call subplots, we think of our setup as an array of plots. It'll have some $i$ number of rows and some $j$ number of columns. Whenever we use plt.subplot( ), we have to declare the number of rows and columns, as well as which element of that total array we're plotting into.

For each subplot, we have to include all the little bits we did before to customize our plot.

```python
dticks = 30

i = 2 # rows
j = 2 # columns

plt.figure(figsize=(10,6))

ax = plt.subplot(i,j,1)
my_xticks = np.arange(-180, 0+dticks, dticks)

ax.set_title('My Galactic Latitude - Longitude Plot')
ax.scatter(gal_long, gal_lat, s=1, c='blue', edgecolor='None')
ax.text(0.03,0.9,'Plot 1',transform=ax.transAxes)

ax.set_xlabel('Galactic Longitude (degrees)')
ax.set_ylabel('Galactic Latitude (degrees)')
ax.set_xlim(-180,0)
ax.set_ylim(0,90)
ax.set_xticks(my_xticks)
ax.minorticks_on()
ax.grid()

ax = plt.subplot(i,j,2)
my_xticks = np.arange(0, 180+dticks, dticks)

ax.set_title('My Galactic Latitude - Longitude Plot')
ax.scatter(gal_long, gal_lat, s=1, c='blue', edgecolor='None')
ax.text(0.03,0.9,'Plot 2',transform=ax.transAxes)

ax.set_xlabel('Galactic Longitude (degrees)')
ax.set_ylabel('Galactic Latitude (degrees)')
ax.set_xlim(0,180)
ax.set_ylim(0,90)
ax.set_xticks(my_xticks)
ax.minorticks_on()
ax.grid()

ax = plt.subplot(i,j,3)
my_xticks = np.arange(-180, 0+dticks, dticks)

ax.set_title('My Galactic Latitude - Longitude Plot')
ax.scatter(gal_long, gal_lat, s=1, c='blue', edgecolor='None')
ax.text(0.03,0.03,'Plot 3',transform=ax.transAxes)

ax.set_xlabel('Galactic Longitude (degrees)')
ax.set_ylabel('Galactic Latitude (degrees)')
ax.set_xlim(-180,0)
ax.set_ylim(-90,0)
ax.set_xticks(my_xticks)
ax.minorticks_on()
ax.grid()

ax = plt.subplot(i,j,4)
my_xticks = np.arange(0, 180+dticks, dticks)

ax.set_title('My Galactic Latitude - Longitude Plot')
ax.scatter(gal_long, gal_lat, s=1, c='blue', edgecolor='None')
ax.text(0.03,0.03,'Plot 4',transform=ax.transAxes)

ax.set_xlabel('Galactic Longitude (degrees)')
ax.set_ylabel('Galactic Latitude (degrees)')
ax.set_xlim(0,180)
ax.set_ylim(-90,0)
ax.set_xticks(my_xticks)
ax.minorticks_on()
ax.grid()
```

EW! First of all, we can easily smash these axes together, since they're all plotting some portion of the same data. We do this by changing the way we initialize the overall figure. Instead of plt.figure( ), we use fig = plt.figure( ), and then adjust the distribution of plots using fig.subplots_adjust( ). If we want to control the horizontal space between plots, we use the keyword "wspace".  For the vertical space, we use "hspace". To smash the plots together so that they share axes, we set both to zero.

We also don't need to add a title to every subplot now.  We can use plt.suptitle( ) to use one overall title.

```python
dticks = 30

i = 2 # rows
j = 2 # columns

fig = plt.figure(figsize=(10,6))
fig.subplots_adjust( hspace=0, wspace=0 )
plt.suptitle('My Galactic Latitude - Longitude Plot')

ax = plt.subplot(i,j,1)
my_xticks = np.arange(-180, 0+dticks, dticks)

ax.scatter(gal_long, gal_lat, s=1, c='blue', edgecolor='None')
ax.text(0.03,0.9,'Plot 1',transform=ax.transAxes)

ax.set_xlabel('Galactic Longitude (degrees)')
ax.set_ylabel('Galactic Latitude (degrees)')
ax.set_xlim(-180,0)
ax.set_ylim(0,90)
ax.set_xticks(my_xticks)
ax.minorticks_on()
ax.grid()

ax = plt.subplot(i,j,2)
my_xticks = np.arange(0, 180+dticks, dticks)

ax.scatter(gal_long, gal_lat, s=1, c='blue', edgecolor='None')
ax.text(0.03,0.9,'Plot 2',transform=ax.transAxes)

ax.set_xlabel('Galactic Longitude (degrees)')
ax.set_ylabel('Galactic Latitude (degrees)')
ax.set_xlim(0,180)
ax.set_ylim(0,90)
ax.set_xticks(my_xticks)
ax.minorticks_on()
ax.grid()

ax = plt.subplot(i,j,3)
my_xticks = np.arange(-180, 0+dticks, dticks)

ax.scatter(gal_long, gal_lat, s=1, c='blue', edgecolor='None')
ax.text(0.03,0.03,'Plot 3',transform=ax.transAxes)

ax.set_xlabel('Galactic Longitude (degrees)')
ax.set_ylabel('Galactic Latitude (degrees)')
ax.set_xlim(-180,0)
ax.set_ylim(-90,0)
ax.set_xticks(my_xticks)
ax.minorticks_on()
ax.grid()

ax = plt.subplot(i,j,4)
my_xticks = np.arange(0, 180+dticks, dticks)

ax.scatter(gal_long, gal_lat, s=1, c='blue', edgecolor='None')
ax.text(0.03,0.03,'Plot 4',transform=ax.transAxes)

ax.set_xlabel('Galactic Longitude (degrees)')
ax.set_ylabel('Galactic Latitude (degrees)')
ax.set_xlim(0,180)
ax.set_ylim(-90,0)
ax.set_xticks(my_xticks)
ax.minorticks_on()
ax.grid()
```

It's still a little ugly.  We need to make it so that the y-axes of the rightmost plots don't overlap the actual plotting area of the leftmost plots.  To that we have to switch up the order in which we declare each plot.  Declare the rightmost one in each row first. That way, each next subplot declared will cover up that y-axis.

```python
dticks = 30

i = 2 # rows
j = 2 # columns

fig = plt.figure(figsize=(10,6))
fig.subplots_adjust( hspace=0, wspace=0 )
plt.suptitle('My Galactic Latitude - Longitude Plot')

ax = plt.subplot(i,j,2)
my_xticks = np.arange(0, 180+dticks, dticks)

ax.scatter(gal_long, gal_lat, s=1, c='blue', edgecolor='None')
ax.text(0.03,0.9,'Plot 2',transform=ax.transAxes)

ax.set_xlabel('Galactic Longitude (degrees)')
ax.set_ylabel('Galactic Latitude (degrees)')
ax.set_xlim(0,180)
ax.set_ylim(0,90)
ax.set_xticks(my_xticks)
ax.minorticks_on()
ax.grid()

ax = plt.subplot(i,j,1)
my_xticks = np.arange(-180, 0+dticks, dticks)

ax.scatter(gal_long, gal_lat, s=1, c='blue', edgecolor='None')
ax.text(0.03,0.9,'Plot 1',transform=ax.transAxes)

ax.set_xlabel('Galactic Longitude (degrees)')
ax.set_ylabel('Galactic Latitude (degrees)')
ax.set_xlim(-180,0)
ax.set_ylim(0,90)
ax.set_xticks(my_xticks)
ax.minorticks_on()
ax.grid()


ax = plt.subplot(i,j,4)
my_xticks = np.arange(0, 180+dticks, dticks)

ax.scatter(gal_long, gal_lat, s=1, c='blue', edgecolor='None')
ax.text(0.03,0.03,'Plot 4',transform=ax.transAxes)

ax.set_xlabel('Galactic Longitude (degrees)')
ax.set_ylabel('Galactic Latitude (degrees)')
ax.set_xlim(0,180)
ax.set_ylim(-90,0)
ax.set_xticks(my_xticks)
ax.minorticks_on()
ax.grid()

ax = plt.subplot(i,j,3)
my_xticks = np.arange(-180, 0+dticks, dticks)

ax.scatter(gal_long, gal_lat, s=1, c='blue', edgecolor='None')
ax.text(0.03,0.03,'Plot 3',transform=ax.transAxes)

ax.set_xlabel('Galactic Longitude (degrees)')
ax.set_ylabel('Galactic Latitude (degrees)')
ax.set_xlim(-180,0)
ax.set_ylim(-90,0)
ax.set_xticks(my_xticks)
ax.minorticks_on()
ax.grid()
```

We're almost there! The last thing we need to do is clean up those little numbers peaking out from behind the big plot. We just do that by adjusting our ylims and xlims slightly. The amount by which you adjust your limits will of course depend on the scale of your axis.  For something on the order of hundred, adjusting by 1 should be sufficient.

<b>Note</b>: if you use plt.xlim or ax.set_xlim after you declare your tickmarks with xticks or ax.set_xticks, then you can easily adjust your axes.

```python
dticks = 30

i = 2 # rows
j = 2 # columns

fig = plt.figure(figsize=(10,6))
fig.subplots_adjust( hspace=0, wspace=0 )
plt.suptitle('My Galactic Latitude - Longitude Plot')

ax = plt.subplot(i,j,2)
my_xticks = np.arange(0, 180+dticks, dticks)

ax.scatter(gal_long, gal_lat, s=1, c='blue', edgecolor='None')
ax.text(0.03,0.9,'Plot 2',transform=ax.transAxes)

ax.set_xlabel('Galactic Longitude (degrees)')
ax.set_ylabel('Galactic Latitude (degrees)')
ax.set_xticks(my_xticks)
ax.set_xlim(0,179)
ax.set_ylim(0,89)
ax.minorticks_on()
ax.grid()

ax = plt.subplot(i,j,1)
my_xticks = np.arange(-180, 0+dticks, dticks)

ax.scatter(gal_long, gal_lat, s=1, c='blue', edgecolor='None')
ax.text(0.03,0.9,'Plot 1',transform=ax.transAxes)

ax.set_xlabel('Galactic Longitude (degrees)')
ax.set_ylabel('Galactic Latitude (degrees)')
ax.set_xticks(my_xticks)
ax.set_xlim(-179,0)
ax.set_ylim(0,89)
ax.minorticks_on()
ax.grid()


ax = plt.subplot(i,j,4)
my_xticks = np.arange(0, 180+dticks, dticks)

ax.scatter(gal_long, gal_lat, s=1, c='blue', edgecolor='None')
ax.text(0.03,0.03,'Plot 4',transform=ax.transAxes)

ax.set_xlabel('Galactic Longitude (degrees)')
ax.set_ylabel('Galactic Latitude (degrees)')
ax.set_xticks(my_xticks)
ax.set_xlim(0,179)
ax.set_ylim(-89,0)
ax.minorticks_on()
ax.grid()

ax = plt.subplot(i,j,3)
my_xticks = np.arange(-180, 0+dticks, dticks)

ax.scatter(gal_long, gal_lat, s=1, c='blue', edgecolor='None')
ax.text(0.03,0.03,'Plot 3',transform=ax.transAxes)

ax.set_xlabel('Galactic Longitude (degrees)')
ax.set_ylabel('Galactic Latitude (degrees)')
ax.set_xticks(my_xticks)
ax.set_xlim(-179,0)
ax.set_ylim(-89,0)
ax.minorticks_on()
ax.grid()
```

And there you have it! Now you have a nice, big, beautiful plot!

### Saving Your Plot

Now that you've gone and made your big, beautiful plot, you might want to actually save that plot somewhere other than your iPython notebook.  Well, that's very simply done with plt.savefig( ).  With plt.savefig, all you have to do is provide your output file name (complete with format), and savefig does the rest.

If you don't provide a path as part of your file name (e.g. fname = 'awesomeplot.pdf'), it'll save to whatever your current directory happens to be. If you do provide a path (e.g. fname = '/my/entire/path/here/lovemyplot.png'), it'll save the file to that path. Here's how you do it.

```python
dticks = 30

i = 2 # rows
j = 2 # columns

fig = plt.figure(figsize=(10,6))
fig.subplots_adjust( hspace=0, wspace=0 )
plt.suptitle('My Galactic Latitude - Longitude Plot')

ax = plt.subplot(i,j,2)
my_xticks = np.arange(0, 180+dticks, dticks)

ax.scatter(gal_long, gal_lat, s=1, c='blue', edgecolor='None')
ax.text(0.03,0.9,'Plot 2',transform=ax.transAxes)

ax.set_xlabel('Galactic Longitude (degrees)')
ax.set_ylabel('Galactic Latitude (degrees)')
ax.set_xticks(my_xticks)
ax.set_xlim(0,179)
ax.set_ylim(0,89)
ax.minorticks_on()
ax.grid()

ax = plt.subplot(i,j,1)
my_xticks = np.arange(-180, 0+dticks, dticks)

ax.scatter(gal_long, gal_lat, s=1, c='blue', edgecolor='None')
ax.text(0.03,0.9,'Plot 1',transform=ax.transAxes)

ax.set_xlabel('Galactic Longitude (degrees)')
ax.set_ylabel('Galactic Latitude (degrees)')
ax.set_xticks(my_xticks)
ax.set_xlim(-179,0)
ax.set_ylim(0,89)
ax.minorticks_on()
ax.grid()


ax = plt.subplot(i,j,4)
my_xticks = np.arange(0, 180+dticks, dticks)

ax.scatter(gal_long, gal_lat, s=1, c='blue', edgecolor='None')
ax.text(0.03,0.03,'Plot 4',transform=ax.transAxes)

ax.set_xlabel('Galactic Longitude (degrees)')
ax.set_ylabel('Galactic Latitude (degrees)')
ax.set_xticks(my_xticks)
ax.set_xlim(0,179)
ax.set_ylim(-89,0)
ax.minorticks_on()
ax.grid()

ax = plt.subplot(i,j,3)
my_xticks = np.arange(-180, 0+dticks, dticks)

ax.scatter(gal_long, gal_lat, s=1, c='blue', edgecolor='None')
ax.text(0.03,0.03,'Plot 3',transform=ax.transAxes)

ax.set_xlabel('Galactic Longitude (degrees)')
ax.set_ylabel('Galactic Latitude (degrees)')
ax.set_xticks(my_xticks)
ax.set_xlim(-179,0)
ax.set_ylim(-89,0)
ax.minorticks_on()
ax.grid()

fname = 'awesomeplot.pdf'
plt.savefig(fname)
```
